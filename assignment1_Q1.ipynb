```python
import matplotlib.pyplot as plt
import numpy as np
def show(name, n, m, i, Title):
    plt.subplot(n, m, i)
    plt.imshow(name, cmap='gray')
    plt.title(Title)
    plt.axis('off')
img1 = np.random.rand(100, 100)
img2 = np.random.rand(100, 100)
plt.figure(figsize=(10, 5))  # Set the figure size

show(img1, 1, 2, 1, 'Random Image 1')  # First image at position 1
show(img2, 1, 2, 2, 'Random Image 2')  # Second image at position 2

plt.show()
```
